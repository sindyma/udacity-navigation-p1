This notebook trains an agent to solve the Banana environment as project 1 of the [Deep Reinforcement Learning Nanodegree](https://www.udacity.com/course/deep-reinforcement-learning-nanodegree--nd893).

### 1. Start the Environment

We begin by importing some necessary packages.  If the code cell below returns an error, please revisit the project instructions to double-check that you have installed [Unity ML-Agents](https://github.com/Unity-Technologies/ml-agents/blob/master/docs/Installation.md) and [NumPy](http://www.numpy.org/).

```python
from unityagents import UnityEnvironment
import numpy as np
import datetime
import torch
from collections import deque
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
env = UnityEnvironment(file_name="Banana.app")
```

Environments contain **_brains_** which are responsible for deciding the actions of their associated agents. Here we check for the first brain available, and set it as the default brain we will be controlling from Python.

```python
# get the default brain
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
```

### 2. Examine the State and Action Spaces

The simulation contains a single agent that navigates a large environment.  At each time step, it has four actions at its disposal:
- `0` - walk forward 
- `1` - walk backward
- `2` - turn left
- `3` - turn right

The state space has `37` dimensions and contains the agent's velocity, along with ray-based perception of objects around agent's forward direction.  A reward of `+1` is provided for collecting a yellow banana, and a reward of `-1` is provided for collecting a blue banana. 

Run the code cell below to print some information about the environment.

```python
# reset the environment
env_info = env.reset(train_mode=False)[brain_name]

# number of agents in the environment
print('Number of agents:', len(env_info.agents))

# number of actions
action_size = brain.vector_action_space_size
print('Number of actions:', action_size)

# examine the state space 
state = env_info.vector_observations[0]
print('States look like:', state)
state_size = len(state)
print('States have length:', state_size)
```

### 4. Train the agent to solve the environment.

This next section of cells trains the agent to solve the environment. Note that we import Agent from ``dqn_agent.py``

```python
from dqn_agent import Agent

agent = Agent(state_size=state_size, action_size=action_size, seed=0, fc1=64, fc2=32)
```

```python
def dqn(n_episodes, eps_start, eps_end, eps_decay):
    scores = []                        # list containing scores from each episode
    window_size = 100
    scores_window = deque(maxlen=window_size)  # last window of scores
    eps = eps_start                    # initialize epsilon
    for i_episode in range(1, n_episodes+1):
        env_info = env.reset(train_mode=True)[brain_name]
        state = env_info.vector_observations[0]
        score = 0
        while True:
            action = agent.act(state, eps) #take an action from the agent
            ###Environment moves a step
            env_info = env.step(action)[brain_name]        # send the action to the environment
            next_state = env_info.vector_observations[0]   # get the next state
            reward = env_info.rewards[0]                   # get the reward
            done = env_info.local_done[0]                  # see if episode has finished
            score += reward                                # update the score
            ###Agent moves a step
            agent.step(state, action, reward, next_state, done)
            #for the next time step
            state = next_state                             # roll over the state to next time step
            
            
            if done:
                break 
        scores_window.append(score)       # save most recent score
        scores.append(score)              # save most recent score
        if i_episode > 100:
            eps = max(eps_end, eps_decay*eps) # decrease epsilon
        #Print statements at different points in time
        if i_episode % window_size == 0:
            currentDT = datetime.datetime.now()
            print('\r{} Episode {}\tAverage Score: {:.2f}\tEpsilon: {:.2f} Loss {:.2f}'.format(currentDT.strftime("%Y-%m-%d %H:%M:%S"),i_episode, np.mean(scores_window),eps,sum(agent.losses[-100:])/100.))
        if i_episode == n_episodes:
            print('\nFinished {:d} episodes\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
            finished_save_name = 'solvingbananas.pth'
            torch.save(agent.qnetwork_local.state_dict(), finished_save_name)
            break
    return scores
```

```python
#Set hyperparameters
n_episodes=600           #Number of episodes to train for
eps_start=1.0             #Where epsilon starts
eps_end=0.01              #Where epsilon ends
eps_decay=0.99         #Epsilon decay rate, dictates exploration vs exploitation rate
currentDT = datetime.datetime.now()
print ("Starting training:",currentDT.strftime("%Y-%m-%d %H:%M:%S"))
scores = dqn(n_episodes, eps_start, eps_end, eps_decay)
# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
```

```python
# Plot the losses
import pandas as pd
loss_pd = pd.DataFrame(agent.losses)
loss_pd.rolling(100).mean().loc[range(0, len(loss_pd), 100), :].plot(figsize=(12, 4))
```

### 5. Run a trained agent

```python
# load the weights from file
agent.qnetwork_local.load_state_dict(torch.load('solvingbananas.pth'))
run_for = 100
scores = []
for i_ep in range(1,run_for+1):
    env_info = env.reset(train_mode=True)[brain_name]
    state = env_info.vector_observations[0]
    score = 0
    while True:
        action = agent.act(state) #take an action from the agent
        env_info = env.step(action)[brain_name]        # send the action to the environment
        next_state = env_info.vector_observations[0]   # get the next state
        reward = env_info.rewards[0]                   # get the reward
        done = env_info.local_done[0]                  # see if episode has finished
        score += reward                                # update the score
        state = next_state                             # roll over the state to next time step
        if done:                                       # exit loop if episode finished
            break
    
    scores.append(score)

print("Avg: {}".format(np.mean(scores)))
# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
```
